```python
X = df[['MedInc']]
y = df['MedHouseVal']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.75, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
model1 = linear.fit(X_train_scaled, y_train)
y_pred = model1.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
adj_r2 = 1 - (1 - r2)*(len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1)
print(f"mean_squared_error : {mse}")
print(f"mean_absolute_error: {mae}")
print(f"R2 score: {r2}")
print(f"Adjusted R2 score: {adj_r2}")
```

**NOTE** : here accuracy is only 59 % and it is too low, so we need to do feature engineering things and increaser our efficiency:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example for features
for col in X.columns:
    plt.figure()
    sns.histplot(X[col], kde=True)
    plt.title(col)
    plt.show()
```

```python
import numpy as np

skewed_features = ['MedInc', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup']

for col in skewed_features:
    X[col + '_log'] = np.log(X[col] + 1)  # +1 avoids log(0)
```

```python
for col in X.columns:
    if '_log' in col:  # check if column name contains '_log'
        plt.figure()
        sns.histplot(X[col], kde=True)
        plt.title(col)
        plt.show()
```

```python
for col in X.columns:
    if '_log' in col or col == 'Longitude' or col == 'Latitude':
        plt.figure()
        plt.boxplot(X[col])
        plt.title(col)
        plt.show()
```

```python
X.columns
```

```python
X.drop(['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup'], axis=1, inplace=True)
```

```python
X.columns
```

```python
for col in X.columns:
    q1 = np.quantile(X[col], 0.25)   # 25th percentile
    q3 = np.quantile(X[col], 0.75)   # 75th percentile
    IQR = q3 - q1
    l = q1 - 1.5 * IQR               # lower bound
    h = q3 + 1.5 * IQR               # upper bound
    
    print(f"----- {col} -----")
    count = len(X[(X[col] >= l) & (X[col] <= h)])   # use '&'
    outliers = len(X) - count
    print(f"Count within bounds: {count}")
    print(f"Outliers: {outliers}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, random_state=42)
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
linear = LinearRegression()
model2 = linear.fit(X_train_scaled, y_train)
y_pred = model2.predict(X_test_scaled)
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
adj_r2 = 1 - (1 - r2)*(len(X_test) - 1) / (len(X_test) - X_test.shape[1] - 1)
print(f"mean_squared_error: {mse}")
print(f"mean_absolute_error: {mae}")
print(f"Root mean squred error: {rmse}")
print(f"r2 score: {r2}")
print(f"Adjusted R2 score: {adj_r2}")
```

**NOTE** Model imporves by only ~2 percent (59 to 61)

```python
from sklearn.linear_model import Lasso, LassoCV

lasso = Lasso()
model3 = lasso.fit(X_train_scaled, y_train)
```

```python
y_pred = model3.predict(X_test_scaled)
```

```python
from sklearn.linear_model import LassoCV

# LassoCV will test multiple alpha values and pick the best one
lasso_cv = LassoCV(alphas=np.logspace(-4, 4, 50), cv=5, random_state=42)
lasso_cv.fit(X_train_scaled, y_train)

print("Best alpha:", lasso_cv.alpha_)
print("Train R2:", lasso_cv.score(X_train_scaled, y_train))
print("Test R2:", lasso_cv.score(X_test_scaled, y_test))
```

```python
# Coefficients (feature importance)
coef = pd.Series(lasso_cv.coef_, index=X.columns)
print("\nLasso coefficients:")
print(coef)

# Features dropped (coeff = 0)
dropped = coef[coef == 0].index.tolist()
kept = coef[coef != 0].index.tolist()

print("\nDropped features:", dropped)
print("Kept features:", kept)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=3, include_bias=False)
X_train_poly = poly.fit_transform(X_train_scaled)
X_test_poly = poly.transform(X_test_scaled)

from sklearn.linear_model import LinearRegression
lin_poly = LinearRegression()
lin_poly.fit(X_train_poly, y_train)

print("Poly Train R2:", lin_poly.score(X_train_poly, y_train))
print("Poly Test R2:", lin_poly.score(X_test_poly, y_test))
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline

ridge_poly = make_pipeline(PolynomialFeatures(degree=3), Ridge(alpha=1.0))
ridge_poly.fit(X_train_scaled, y_train)

print("Ridge Poly Train R2:", ridge_poly.score(X_train_scaled, y_train))
print("Ridge Poly Test R2:", ridge_poly.score(X_test_scaled, y_test))
```
